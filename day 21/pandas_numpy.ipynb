```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Department': ['HR', 'IT', 'HR', 'IT', 'HR', 'IT'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'F'],
    'Salary': [50000, 70000, 55000, 80000, 52000, 75000]
}
```

```python
df = pd.DataFrame(data)
result = df.groupby(['Department', 'Gender'])['Salary'].sum()
print(result)
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'HR', 'IT', 'HR'],
    'Salary': [50000, 70000, 55000, 80000, 52000]
}

df = pd.DataFrame(data)

result = pd.pivot_table(
    df,
    values='Salary',
    index='Department',
    aggfunc='mean'
)

print(result)
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Department': ['HR', 'IT', 'HR', 'IT', 'HR', 'IT'],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'F'],
    'Salary': [50000, 70000, 55000, 80000, 52000, 75000]
}

df = pd.DataFrame(data)

result = pd.pivot_table(
    df,
    values='Salary',
    index='Department',
    columns='Gender',
    aggfunc='sum'
)

print(result)
```

```python
data = {'Sales': [100, 200, 300, 400]}
df = pd.DataFrame(data)
df['Previous Sales'] = df['Sales'].shift(1)
print(df)
```

```python
df['Rank'] = df['Sales'].rank()
print(df)
```

```python
df['Rolling Mean'] = df['Sales'].rolling(window=2).mean()
print(df)
```
