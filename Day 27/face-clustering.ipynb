```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import cv2
import os

print("Loading image and applying DBSCAN clustering...\n")

# 1. Load image and detect points (Example using simple edges)
image_file = 'face.jpg'

if not os.path.exists(image_file):
    print(f"Error: {image_file} not found!")
    print("Available files:", os.listdir('.'))
    exit(1)

image = cv2.imread(image_file, 0)

if image is None:
    print(f"Error: Could not read {image_file}")
    exit(1)

print(f"✓ Image loaded: {image_file}")
print(f"  Image shape: {image.shape}\n")

# Apply edge detection
edges = cv2.Canny(image, 100, 200)
points = np.column_stack(np.where(edges > 0)) # Get (y, x) coordinates

print(f"✓ Edge detection completed")
print(f"  Points detected: {len(points)}\n")

# 2. Apply DBSCAN
# eps: distance between points; min_samples: density threshold
print("Applying DBSCAN clustering...")
db = DBSCAN(eps=10, min_samples=10).fit(points)
labels = db.labels_

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"✓ DBSCAN completed")
print(f"  Clusters found: {n_clusters}")
print(f"  Noise points: {n_noise}\n")

# 3. Visualize the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Face Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.scatter(points[:, 1], points[:, 0], c=labels, cmap='viridis', s=1)
plt.title(f"Face Features Segmented by DBSCAN\n({n_clusters} clusters detected)")
plt.gca().invert_yaxis() # Images have inverted Y-axis
plt.axis('off')

plt.tight_layout()

# Save the plot instead of showing it
output_file = 'dbscan_face_segmentation.png'
plt.savefig(output_file, dpi=100, bbox_inches='tight')
print(f"✓ Visualization saved: {output_file}")
plt.close()
```
